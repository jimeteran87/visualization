```python

!pip install openpyxl
import pandas as pd
# Load Excel file
df = pd.read_excel("bikeshare-ridership-2016.xlsx")

# Show first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Count trips by start station
top_stations = df['from_station_name'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
top_stations.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 10 Start Stations - Bike Share Toronto (2016)')
plt.xlabel('Number of Trips')
plt.ylabel('Start Station')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Select relevant columns for the Excel visualization
df_cleaned = df[['trip_id', 'trip_start_time', 'trip_stop_time', 'trip_duration_seconds',
                 'from_station_name', 'to_station_name', 'user_type']]

# Export the cleaned dataset to a CSV file
df_cleaned.to_csv("cleaned_bike_ridership_2016.csv", index=False)
```
