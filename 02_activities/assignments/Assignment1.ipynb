```python
#July 9th

#Multiple Viz on one axes objcet


# sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array ([110, 180, 240, 99, 220])
y2 = np.array ([170, 100, 90, 120, 50])
# figure and axes 
fig, ax = plt.subplots(figsize=(7, 3))
# now we call both bar and plot elements to the same axes (ax)
ax.bar (x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
#Adding error information 
y2_sd = np.std (y2)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x,y2,
yerr = y2_sd,
fmt = "none")
```

```python
#Customize errorbar

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x,y2,
yerr = y2_sd,
fmt = "none",
ecolor = "indigo",
elinewidth = 6.
capsize=6,
capthick=4)
```

```python
# Using images
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([1170, 100, 90, 120, 501])

# Get image
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)

# Plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')

# Add image as a new axes on top of the plot
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis('off')  # Hide axes for the image

plt.show()
```

```python
#July 10th
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Set random seed for reproducibility
np.random.seed(613)

#load data
tips = sns.load_dataset("tips")
print (tips)


# Line plot
sns.set_style('dark')
tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip')

# Add title and labels
tipgraph.set(
    title='Tips vs. Total Bill',
    ylabel='Tip Amount ($)',
    xlabel='Total Bill ($)'
)

#aesthetic objects
tipgraph = sns.lineplot(
    data=tips,
    x='total_bill',
    y='tip',
    color='hotpink',
    linestyle='--',
    linewidth=3,
    marker='o',
    markerfacecolor='indigo'
)


plt.show()
```

```python
# Scatter plot


# Custom palette (works in all seaborn versions)
custom_palette = {
    'Thur': 'purple',
    'Fri': 'hotpink',
    'Sat': 'deepskyblue',
    'Sun': 'yellowgreen'
}

# Scatter plot
tipgraph = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    style='time',    # Different markers for Lunch/Dinner
    hue='day',       # Color by day
    palette=custom_palette,
    s=100,           # Make points larger
    edgecolor='black'  # Add black edge to points
)

# Add title and labels
tipgraph.set(
    title='Tips vs. Total Bill',
    xlabel='Total Bill ($)',
    ylabel='Tip Amount ($)'
)

# Show plot
plt.show()
```

```python
# Plotly

import plotly.graph_objects as go
import numpy as np

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])

graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(
    title='Pirate Scores',
    xaxis_title='Pirates',
    yaxis_title='Score'
)

graph.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])

graph = go.Figure()
graph.add_trace(go.Scatter(
    x=x1, 
    y=y1,
    mode='markers',
    marker=dict(
        size=15,
        color='hotpink',
        opacity=1,
        line=dict(width=5, color='purple')
    )
))

graph.update_layout(
    title='Pirate Scores',
    xaxis_title='Pirates',
    yaxis_title='Score',
    width=500,
    height=500
)

graph.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Unir todas las frases en un solo string
text = " ".join(each for each in df.quote.astype(str))

# Crear la nube de palabras
wordcloud = WordCloud(
    background_color='white',
    colormap='inferno'
).generate(text)

# Mostrar la nube de palabras
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis('off')
plt.show()
```
